# Defining Functions

#### Function: dataset_get_report_refresh

##### Purpose

To identify a Power BI report's linked dataset's refresh. 

This is function uses following API: [Datasets - Get Refresh Schedule](https://learn.microsoft.com/en-us/rest/api/power-bi/datasets/get-refresh-schedule)
Then the Semantic Link Labs Functions: get_semantic_model_refresh_history(), resolve_dataset_from_report()

##### Parameters
- report : str | uuid.UUID
    - The name or ID of the Power BI report.
- workspace : str | uuid.UUID, default=None
    -  The Fabric workspace name or ID in which the report exists. Defaults to None which resolves to the workspace of the attached lakehouse or if no lakehouse attached, resolves to the workspace of the notebook.

##### Returns 

All date times are in UTC. 

A Tuple: 
- last_refresh_type: The last refresh type
- last_refresh_status: The status of the last refresh
- last_refresh_start: The start time of the last refresh - if null returns July 11th, 2011
- last_refresh_end: The end time of the last refresh - if null returns July 11th, 2011
- next_refresh_date: The next scheduled refresh of the power bi report - if null returns July 11th, 2011
- rpt_refresh_enabled: returns a boolean if the refresh is enabled or not (True/false)

#### Function: get_last_refresh_status

##### Purpose

To loop thru the reports in a workspace and calls the function dataset_get_report_refresh for every single report in the workspace. 

##### Parameters
- report_workspace : str
    - The name of a Power BI workspace

##### Returns 

a Pandas data frame with the results of calling dataset_get_report_refresh for every single report.


```python
from typing import Tuple,  Optional
from uuid import UUID
from datetime import datetime, timezone, timedelta 
import pandas as pd
import sempy_labs as sl
import sempy.fabric as fabric
from sempy_labs._helper_functions import (
    _base_api
    , resolve_dataset_from_report
)
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, TimestampType, BooleanType

def dataset_get_report_refresh(
    report: str | UUID
    , report_workspace: Optional[str | UUID] = None
) -> Tuple[str, str, datetime, datetime, datetime, bool]:

    #setting defaults
    last_refresh_type = "Not Refreshed"
    last_refresh_status = "Not Refreshed"
    last_refresh_start = datetime(2011, 7, 11, tzinfo=timezone.utc)
    last_refresh_end = datetime(2011, 7, 11, tzinfo=timezone.utc)
    next_refresh = datetime(2011, 7, 11, tzinfo=timezone.utc)
    rpt_refresh_enabled = False
    error_tuple = (last_refresh_type, 'Failed to Check Refresh Status', last_refresh_start, last_refresh_end, datetime(2011, 7, 11, tzinfo=timezone.utc), rpt_refresh_enabled)
    error_message = f"Report: {report} in Workspace: {report_workspace} failed to retrive refresh"

    #using semantic link to get the reports dataset
    try:
        dataset_tuple = resolve_dataset_from_report(report=report, workspace=report_workspace)
    except:
        print(error_message)
        return error_tuple

    dataset_id = dataset_tuple[0]
    dataset_workspace = dataset_tuple[2]
    
    try:
        last_refresh = sl.get_semantic_model_refresh_history(
            dataset= dataset_id
            , workspace = dataset_workspace
        ).sort_values(
            by='Start Time'
            , ascending=False
        ).head(1)
    except:
        print(error_message)
        return error_tuple

    if not last_refresh.empty:
        last_refresh_type = last_refresh.iloc[0]['Refresh Type']
        last_refresh_status = last_refresh.iloc[0]['Status']
        last_refresh_start = datetime.strptime(last_refresh.iloc[0]['Start Time'], "%Y-%m-%dT%H:%M:%S.%fZ")
        last_refresh_end = datetime.strptime(last_refresh.iloc[0]['End Time'], "%Y-%m-%dT%H:%M:%S.%fZ")

    #calling the power bi api to get the refresh schedule
    try:
        response = _base_api(
            request=f"v1.0/myorg/datasets/{dataset_id}/refreshSchedule"
        )
    except:
        print(error_message)
        return error_tuple
    
    #moving api results into local variables
    rpt_refresh_timezone = response.json()['localTimeZoneId']  
    rpt_refresh_days = response.json()['days']
    rpt_refresh_times = response.json()['times']
    rpt_refresh_enabled = response.json()["enabled"]

    #If refresh is not enabled return July 11th 2011
    if not rpt_refresh_enabled:
        return (last_refresh_type, last_refresh_status, last_refresh_start, last_refresh_end, datetime(2011, 7, 11, tzinfo=timezone.utc), rpt_refresh_enabled)

    #caclulates a timezoneoffset using the dictionary based off the localTimeZoneId PBI API returned
    rpt_refresh_timezone_offset = timezone(timedelta(seconds=timezones[rpt_refresh_timezone]))
    
    #converts the current time into the same timezone that the Power BI dataset is refreshing
    current_utc_date_time = datetime.utcnow()
    current_local_time = current_utc_date_time.astimezone(rpt_refresh_timezone_offset)

    """
    For the day of the week and time the dataset is set to refresh, this calculates the next occurrence of that datetime.

    It uses the current datetime to determine if the day/time combination has yet to occur for a given week.
    - If it has already occurred and it will append the day/time combination for the next week into the array next_occurrences.
    - If it has not already occured it will append the day/time combination for the current week into the array next_occurrences.
    """
    next_occurrences = []    
    for day in rpt_refresh_days:
        for time in rpt_refresh_times:
            target_weekday = day_of_week[day]
            current_weekday = current_local_time.weekday()
            days_until_target = (target_weekday - current_weekday + 7) % 7
            if days_until_target == 0 and current_local_time.time() > datetime.strptime(time, "%H:%M").time():
                days_until_target = 7
            next_date = current_local_time + timedelta(days=days_until_target)
            next_datetime = next_date.replace(hour=int(time.split(":")[0]), minute=int(time.split(":")[1]), second=0, microsecond=0)
            next_occurrences.append(next_datetime)

    #returns the earliest date from the array
    next_refresh_date= min(next_occurrences).astimezone(timezone.utc)
    
    return (last_refresh_type, last_refresh_status, last_refresh_start, last_refresh_end, next_refresh_date, rpt_refresh_enabled)

def get_last_refresh_status(
    report_workspace:str
)-> pd.DataFrame:

    report_type_schema = {
        'Report Id': 'string',
        'Report Name': 'string',
        'Refresh Type': 'string',
        'Last Refresh Status': 'string',
        'Last Refresh Start': 'datetime64[ns]',
        'Last Refresh End': 'datetime64[ns]',
        'Next Refresh Time': 'datetime64[ns]',
        'Next Refresh Enabled':'bool'
    }

    #setting up empty data frame
    report_refresh_df = pd.DataFrame(columns=['Report Id'
    , 'Report Name'
    , 'Refresh Type'
    , 'Last Refresh Status'
    , 'Last Refresh Start'
    , 'Last Refresh End'
    , 'Next Refresh Time'
    , 'Next Refresh Enabled']).astype(report_type_schema)

    #getting a data frame of items in a workspace
    report_df = pd.DataFrame(fabric.list_items(workspace=report_workspace))

    #filtering down to just report item types
    report_df = report_df[report_df['Type'] == 'Report']


    #for each report in dataframe iterrate the rows and call the get report refresh function
    for report in report_df.iterrows():
        refresh_status = dataset_get_report_refresh(report=report[1]['Id'], report_workspace=report[1]['Workspace Id'])
        report_data = {
            'Report Id':report[1]['Id'],
            'Report Name':report[1]['Display Name'],
            'Refresh Type': refresh_status[0],
            'Last Refresh Status': refresh_status[1],
            'Last Refresh Start': pd.to_datetime(refresh_status[2]).tz_localize(None),
            'Last Refresh End': pd.to_datetime(refresh_status[3]).tz_localize(None),
            'Next Refresh Time': pd.to_datetime(refresh_status[4]).tz_localize(None),
            'Next Refresh Enabled': refresh_status[5]
        }
        report_data = pd.DataFrame(report_data,index=['Row1']).astype(report_type_schema)
        #append the row
        report_refresh_df = pd.concat([report_refresh_df, pd.DataFrame(report_data)], ignore_index=True)
    
    #returning the data frame
    return report_refresh_df
```

# Calling Function and Saving Data

```python
report_dataframe = get_last_refresh_status('Production')
display(report_dataframe)

# Define the schema for the table
schema = StructType([
    StructField("report_id", StringType(), True),
    StructField("report_name", StringType(), True),
    StructField("report_type", StringType(), True),
    StructField("last_refresh_status", StringType(), True),
    StructField("last_refresh_start_utc", TimestampType(), True),
    StructField("last_refresh_end_utc", TimestampType(), True),
    StructField("next_Refresh_time_utc", TimestampType(), True),
    StructField("next_refresh_enabled", BooleanType(), True),
])

table_name = 'report_refresh'
df = spark.createDataFrame(report_dataframe, schema)
df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable(table_name)

```
