## Dataset Schema

| Field Name        | Type      | Description                                                                 |
|-------------------|-----------|------------------------------------------------------------------------------|
| configuredBy      | any[]     | List of users who configured the dataset                                    |
| id                | string    | Unique identifier (UUID) of the dataset                                     |
| kind              | string    | Type of refreshable (typically "Dataset")                                   |
| lastRefresh       | any       | Object containing metadata about the last refresh attempt                   |
| name              | string    | Display name of the dataset                                                 |
| refreshSchedule   | any       | Object defining scheduled refresh settings (days, times, timezone, etc.)    |
| averageDuration   | double    | Average duration of refreshes (in seconds)                                  |
| endTime           | string    | End of the refresh history window (UTC)                                     |
| medianDuration    | double    | Median duration of refreshes (in seconds)                                   |
| refreshCount      | bigint    | Total number of refreshes in the time window                                |
| refreshFailures   | bigint    | Number of failed refreshes                                                  |
| refreshesPerDay   | bigint    | Average number of refreshes per day                                         |
| startTime         | string    | Start of the refresh history window (UTC)                                   |
