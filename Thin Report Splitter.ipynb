```python
from packaging.version import Version
from importlib.metadata import version, PackageNotFoundError

try:
    sempy_labs_version = version('semantic-link-labs')
except PackageNotFoundError:
    sempy_labs_version = None

if sempy_labs_version is None or Version(sempy_labs_version) < Version("0.9.8"):
    get_ipython().run_line_magic('pip', 'install semantic-link-labs')
```

```python
import sempy_labs as sl
import sempy_labs.report as slr

start_workspace = 'Production'
report= 'Example_11_Revenue_Target_card'
dataset_name = 'Example_11_Revenue_Target_card'

destination_dataset_workspace = 'Developer'
destination_dataset_name = 'Example_11_Revenue_Target_card Semantic Model'

remove_original_dataset = True
overwrite = True
```

```python
#copying the semantic model
sl.deploy_semantic_model(source_dataset=dataset_name
    , source_workspace=start_workspace
    , target_dataset=destination_dataset_name
    , target_workspace=destination_dataset_workspace
    , refresh_target_dataset=False
    , overwrite=overwrite)

#rebinding the report
slr.report_rebind(report=report
, dataset=destination_dataset_name
, report_workspace=start_workspace
, dataset_workspace=destination_dataset_workspace)

#trying to delete the original dataset
if remove_original_dataset:
    try: 
        sl.delete_semantic_model(dataset=dataset_name, workspace=start_workspace)
    except:
        print("Failed to delete original semantic model. Please go manually delete")
```
