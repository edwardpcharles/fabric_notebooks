### Importing Packages

```python
import sempy.fabric as fabric
import sempy_labs as slr
import pandas as pd
```

### User Inputs

```python
report_workspace_name = 'Sneaker Workspace'
report_name = 'Sales & Returns Sample v201912'
```

### Fetching Workspace and Report Id

```python
workspaces = fabric.list_workspaces()
workspaces = workspaces[workspaces['Name'] == report_workspace_name]
report_workspace_id = workspaces.iloc[0, 0]

report_id = slr.resolve_report_id(workspace=report_workspace_name,report=report_name)

dataset_workspace = slr.resolve_dataset_from_report(workspace=report_workspace_name,report=report_name)

dataset_id = dataset_workspace[0]
dataset_name = dataset_workspace[1]
dataset_workspace_id = dataset_workspace[2]
dataset_workspace_name = dataset_workspace[3]
```

### Fetch all workspace App users (Requires Admin Access)

```python
apps = fabric.list_apps()
apps = apps[apps['Workspace Id'] == report_workspace_id]
for app in apps.itertuples():
    app_id = app[1]
    url = f'/v1.0/myorg/admin/apps/{app_id}/users'
    pbi_client = fabric.PowerBIRestClient()
    users = pbi_client.get(url).json()['value']
    users_df = pd.json_normalize(users)
    display(users_df)
```

### Fetch all Workspace Users

```python
url = f'/v1.0/myorg/groups/{report_workspace_id}/users'
pbi_client = fabric.PowerBIRestClient()
users = pbi_client.get(url).json()['value']
users_df = pd.json_normalize(users)
display(users_df)
```

### Fetch All Report Users (Requires Admin Access)

```python
url = f'/v1.0/myorg/admin/reports/{report_id}/users'
pbi_client = fabric.PowerBIRestClient()
users = pbi_client.get(url).json()['value']
users_df = pd.json_normalize(users)
display(users_df)
```

### Fetch All RLS Roles and Assignments

```python
roles = fabric.get_roles(dataset=dataset_name, workspace=dataset_workspace_name, include_members = True)
display(roles)
```
