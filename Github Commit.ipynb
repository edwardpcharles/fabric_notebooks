```python
def get_git_status(client, workspace_id):
    url = f"https://api.fabric.microsoft.com/v1/workspaces/{workspace_id}/git/status"
    result = client.get(url)
    if result.status_code == 200:
        git_status = result.json()
        return git_status
    else:
        print(f"Error getting Git status: {result.status_code} - {result.text}")
        return None

def commit_to_git(client, workspace_id, workspace_head, commit_message):
    url = f"https://api.fabric.microsoft.com/v1/workspaces/{workspace_id}/git/commitToGit"
    payload = {
        "mode": "All",
        "workspaceHead": workspace_head,
        "comment": commit_message
    }
    headers = {
        "Content-Type": "application/json"
    }
    json_payload = json.dumps(payload)
    result = client.post(url, data=json_payload, headers=headers)
    if result.status_code == 202: 
        return True
    else:
        return False
```
